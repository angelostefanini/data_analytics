```python
# Import packages
import pandas as pd
import seaborn as sns

# Load dataset
penguins = sns.load_dataset("penguins")

# Examine first 5 rows of dataset
penguins.head()
```

```python
print(penguins.columns)
```

```python
# Drop rows with missing values
penguins.dropna(inplace=True)

# Reset index
penguins.reset_index(inplace=True, drop=True)
```

```python
# Subset del DataFrame selezionando solo le colonne di interesse
penguins = penguins[["body_mass_g", "bill_length_mm", "sex", "species"]]

# Rinomina le colonne per una migliore chiarezza
penguins = penguins.rename(columns={"sex": "gender"})
```

```python
# Drop rows with missing values
penguins.dropna(inplace=True)

# Reset index
penguins.reset_index(inplace=True, drop=True)
```

```python
print(penguins)
```

```python
# Subset X and y variables
penguins_X = penguins[["bill_length_mm", "gender", "species"]]
penguins_y = penguins[["body_mass_g"]]
```

```python
# Import train-test-split function from sci-kit learn
from sklearn.model_selection import train_test_split
# Create training data sets and holdout (testing) data sets
X_train, X_test, y_train, y_test = train_test_split(penguins_X, penguins_y,
                                                    test_size = 0.3, random_state = 42)
```

```python
# Write out OLS formula as a string
ols_formula = "body_mass_g ~ bill_length_mm + C(gender) + C(species)"
```

```python
from statsmodels.formula.api import ols
```

```python
# Create OLS dataframe
ols_data = pd.concat([X_train, y_train], axis = 1)

# Create OLS object and fit the model
OLS = ols(formula = ols_formula, data = ols_data)
model = OLS.fit()
```

```python
# Get model results
model.summary()
```

Coefficienti (Coef):
I coefficienti sono valori che rappresentano la variazione attesa nella variabile dipendente per una variazione unitaria nella variabile indipendente corrispondente, mantenendo costanti tutte le altre variabili. Nel tuo modello:

Intercept: Questo è il valore previsto della variabile dipendente quando tutte le variabili indipendenti sono zero. Nel tuo caso, l'intercetta è 2032.2111.

C(gender)[T.Male]: Questo coefficiente è associato alla variabile categorica "gender", con la categoria di riferimento implicita come femmina. Il valore 528.9508 indica la differenza attesa nella variabile dipendente tra i maschi e le femmine, mantenendo costanti le altre variabili.

C(species)[T.Chinstrap]: Questo coefficiente è associato alla variabile categorica "species" con la categoria di riferimento implicita come "Adelie". Il valore -285.3865 indica la differenza attesa nella variabile dipendente tra la specie "Chinstrap" e "Adelie", mantenendo costanti le altre variabili.

C(species)[T.Gentoo]: Questo coefficiente è associato alla variabile categorica "species" con la categoria di riferimento implicita come "Adelie". Il valore 1081.6246 indica la differenza attesa nella variabile dipendente tra la specie "Gentoo" e "Adelie", mantenendo costanti le altre variabili.

bill_length_mm: Questo coefficiente è associato alla variabile continua "bill_length_mm". Il valore 35.5505 indica la variazione attesa nella variabile dipendente per una variazione unitaria nella lunghezza del becco, mantenendo costanti le altre variabili.

P>|t| (Valori p dei t-test):
I valori p dei t-test forniscono informazioni sulla significatività statistica di ciascun coefficiente. I valori p indicano la probabilità di ottenere un t-value almeno tanto estremo quanto quello osservato, supponendo che il coefficiente associato sia in realtà zero (nessun effetto). Nel tuo caso:

P>|t| per tutti i coefficienti: Tutti i valori p sono molto vicini a zero (0.000), il che indica che tutti i coefficienti sono considerati statisticamente significativi. In altre parole, c'è evidenza sufficiente per respingere l'ipotesi nulla che il coefficiente sia zero.
La significatività statistica dei coefficienti è importante per capire quali variabili contribuiscono in modo significativo alla predizione della variabile dipendente nel contesto del tuo modello di regressione lineare.
